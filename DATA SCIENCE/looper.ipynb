```python
with open("cols.txt", "r") as f:
    cols = f.read().split("\n")
cols
```

```python
# template 
values = df['Best all round (Student name)'].str.title().value_counts()

fig, ax = plt.subplots()
values.plot(ax=ax, kind='pie', figsize=(40, 40),autopct=lambda p : '{:.2f}%  ({:,.0f})'.format(p,p * sum(values)/100),
rotatelabels=True, pctdistance=0.9, labeldistance=1.05, 
        wedgeprops = {"edgecolor" : "black",
                      'linewidth': 0.5,
                      'antialiased': True})
ax.set_title("Best All round", fontsize=50)
plt.show()
# fig.savefig("figures/round.png", dpi=300)
```

```python
def pie_chart(column_name,save=False, show=True, title=None, threshold = 0, path="loop_figures/"):
    if title == None:
        title = column_name
    values = df[f'{column_name}'].str.title().value_counts()
    # cut if greater than threshold
    values = values[values >= threshold]

    fig, ax = plt.subplots()
    values.plot(ax=ax, kind='pie', figsize=(20, 20),autopct=lambda p : '{:.2f}%  ({:,.0f})'.format(p,p * sum(values)/100),
    rotatelabels=True, pctdistance=0.9, labeldistance=1.05, ylabel="", 
        wedgeprops = {"edgecolor" : "black",
                      'linewidth': 0.5,
                      'antialiased': True})
    ax.set_title(f"{title}", fontsize=50)
    if show:
        plt.show()
    else:
        plt.close()
    if save:
        fig.savefig(f"{path}{title}.png", dpi=300)

pie_chart('Best all round (Student name)', show=True, threshold=3)
```

```python
count = 0
for col in cols:
    pie_chart(col, save=True, show=False)
    count += 1
    print(f"{col} (loop {count})")
print("completed")
print("completed")
```
